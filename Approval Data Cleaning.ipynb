```python
import pandas as pd
import numpy as np
```

```python
dftrump = pd.read_csv('Approval Ratings- Donald Trump.csv')
```

```python
dftrump.index = pd.to_datetime(dftrump['Start Date'])
```

```python
dftrump = dftrump.drop('Start Date', axis = 1)
```

```python
dftrump['End Date'] = pd.to_datetime(dftrump['End Date'])
```

```python
dftrump = dftrump.resample('M', on='End Date').mean()
```

```python
dftrump['President'] = 'Trump'
```

```python
dftrump.head()
```

```python
dfobama = pd.read_csv('Approval Ratings- Barack Obama.csv')
dfobama.index = pd.to_datetime(dfobama['Start Date'])
dfobama = dfobama.drop('Start Date', axis = 1)
dfobama['End Date'] = pd.to_datetime(dfobama['End Date'])
dfobama = dfobama.resample('M', on='End Date').mean()
dfobama['President'] = 'Obama'
```

```python
dfobama.head()
```

```python
dfwbush = pd.read_csv('Approval Ratings - George W. Bush.csv')
dfwbush.index = pd.to_datetime(dfwbush['Start Date'])
dfwbush = dfwbush.drop('Start Date', axis = 1)
dfwbush['End Date'] = pd.to_datetime(dfwbush['End Date'])
dfwbush = dfwbush.resample('M', on='End Date').mean()
dfwbush['President'] = 'GW Bush'
```

```python
dfwbush.head()
```

```python
dfclinton = pd.read_csv('Approval Ratings- William J. Clinton.csv')
dfclinton.index = pd.to_datetime(dfclinton['Start Date'])
dfclinton = dfclinton.drop('Start Date', axis = 1)
dfclinton['End Date'] = pd.to_datetime(dfclinton['End Date'])
dfclinton = dfclinton.resample('M', on='End Date').mean()
dfclinton['President'] = 'Clinton'
```

```python
frames = [dftrump, dfobama, dfwbush, dfclinton]
dfbig = pd.concat(frames)
```
